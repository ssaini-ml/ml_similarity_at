```python
data = spark.sql("""
                 select * from datascience.at_product_details_individual_embeddings
                 """)
```

```python
display(data.limit(10))
```

```python
from pyspark.sql.functions import countDistinct, count

result = data.agg(
    countDistinct("identifier").alias("distinct_identifier_count"),
    count("identifier").alias("identifier_count"),
    countDistinct("pzn").alias("distinct_pzn_count"),
    count("pzn").alias("pzn_count"),
    countDistinct("upid").alias("distinct_upid_count"),
    count("upid").alias("upid_count")
)

display(result)
```
